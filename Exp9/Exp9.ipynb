```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans

# Load the data
data = pd.read_csv('em9.csv')
X = data[['V1', 'V2']].values

# Plot the whole dataset
plt.title('Graph for the whole dataset')
plt.scatter(X[:, 0], X[:, 1], c='black')
plt.show()

# Apply K-means clustering
kmeans = KMeans(n_clusters=2)
labels_kmeans = kmeans.fit_predict(X)
print("Labels: ", labels_kmeans)
centroids_kmeans = kmeans.cluster_centers_
print("Centroids: ", centroids_kmeans)

# Plot the results using K-means
plt.title('Graph using K-means Algorithm')
plt.scatter(X[:, 0], X[:, 1], c=labels_kmeans)
plt.scatter(centroids_kmeans[:, 0], centroids_kmeans[:, 1], marker='*', c='red')
plt.show()

# Apply Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=2)
labels_gmm = gmm.fit_predict(X)
print("Labels: ", labels_gmm)
means_gmm = gmm.means_
print("Means: ", means_gmm)
covariances_gmm = gmm.covariances_
print("Covariances: ", covariances_gmm)

# Plot the results using GMM
plt.title('Graph using Gaussian Mixture Model (GMM)')
plt.scatter(X[:, 0], X[:, 1], c=labels_gmm)
plt.show()
```
