1. Read a csv file without dataframes

```python
import csv
from pprint import pprint

with open('customer(A).csv', 'r') as f:
    reader = csv.reader(f)
    customer = [row for row in reader]
    pprint(customer)
```

2. Read a csv file using dataframes

```python
import numpy as np
import pandas as pd

df = pd.read_csv('customer(A).csv')
df
```

3. Identify which of the columns and their datatypes

```python
df.dtypes
```

4. Frequency of each unique value in a column

```python
df['income'].value_counts()
```

5. Plot each column graph vertically(bar)

```python
df['income'].value_counts().plot.bar()
```

6. Identify and Delete Rows that Contain Duplicate Data by considering an appropriate

```python
# print duplicate rows and delete them
print("Duplicates:\n",df[df.duplicated()])
print("Result:\n",df.drop_duplicates())
```

7. Identify and Delete Columns That Contain a Single Value by considering a single value

```python
for key in df.keys():
    if(len(df[key].unique()) == 1):
        print(key)
        del df[key]
print(df)
```

8. Transform categorical columns to integer using get_dummies

```python
# Path: Exp3/Exp3.ipynb
transformation = pd.get_dummies(df['income'])
df_new = pd.concat([df, transformation], axis=1)
df_new
```

9. Transform categorical columns to integer using condition

```python
df_new['age'] = [0 if x == 'senior' else 1 for x in df['age']]
df_new
```

10. View top=3 rows

11. View bottom=3 rows

12. Filtering rows and selecting columns by label

```python
df.loc[df['student'] == "yes", ['student','age', 'income']]
```

13. Adding a new instance

```python
# Add a new row
df_new.loc[len(df_new)] = {'age': 0, 'income': 'medium', 'student': 'yes', 'credit_rating': 'fair', 'class': 'yes'}
df_new
```

14. Transform values into bins

```python
new_df = pd.DataFrame({"age_int": np.random.randint(0, 100, 10)})
labels = [f"{i} - {i+9}" for i in range(0, 100, 10)]
new_df["age_group"] = pd.cut(new_df.age_int, range(0, 105, 10), labels=labels)
df_new = pd.concat([df, new_df], axis=1)
df_new.head()
```
